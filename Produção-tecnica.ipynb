```python
import xml.etree.ElementTree as ET
import pandas as pd
import os
import html
```

```python
def processar_arquivo(arquivo_xml):
    try:
        # Tenta analisar o arquivo XML
        tree = ET.parse(arquivo_xml)
        root = tree.getroot()

        # Busca as informações gerais
        dados_gerais = root.find("DADOS-GERAIS")
        nome_completo = html.unescape(dados_gerais.attrib["NOME-COMPLETO"])

        print("Informações dos trabalhos encontrados (Ano >= 2021):")
        print(f"Nome: {nome_completo}")

        trabalhos_Tecnicos = root.findall(".//TRABALHO-TECNICO")
        demais_Apresentacao = root.findall(".//APRESENTACAO-DE-TRABALHO")
        demais_Material = root.findall(".//DESENVOLVIMENTO-DE-MATERIAL-DIDATICO-OU-INSTRUCIONAL")
        demais_editoracao = root.findall(".//EDITORACAO")
        demais_evento = root.findall(".//ORGANIZACAO-DE-EVENTO")
        demais_programa = root.findall(".//PROGRAMA-DE-RADIO-OU-TV")
        demais_relatorio = root.findall(".//RELATORIO-DE-PESQUISA")
        demais_midia = root.findall(".//MIDIA-SOCIAL-WEBSITE-BLOG")
        demais_outra = root.findall(".//OUTRA-PRODUCAO-TECNICA")
        demais_carta = root.findall(".//CARTA-MAPA-OU-SIMILAR")
        demais_curso = root.findall(".//CURSO-DE-CURTA-DURACAO-MINISTRADO")
        


        resultados_trabalho = []
        resultados_demais_apresentacao = []
        resultados_demais_material = []
        resultados_demais_editoracao = []
        resultados_demais_evento = []
        resultados_demais_programa = []
        resultados_demais_relatorio = []
        resultados_demais_midia = []
        resultados_demais_outra = []
        resultados_demais_carta = []
        resultados_demais_curso = []



        ############################################## Processa os trabalhos em eventos

        for trabalho_tecnico in trabalhos_Tecnicos:
            dados_basicos = trabalho_tecnico.find("DADOS-BASICOS-DO-TRABALHO-TECNICO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_trabalho_Tecnico = html.unescape(dados_basicos.attrib.get("TITULO-DO-TRABALHO-TECNICO", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = trabalho_tecnico.find("DETALHAMENTO-DO-TRABALHO-TECNICO")
                    Finalidade = html.unescape(detalhamento.attrib.get("FINALIDADE", "")) if detalhamento is not None else ""
                    Disponibilidade = html.unescape(detalhamento.attrib.get("DISPONIBILIDADE", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE-DO-TRABALHO", "")) if detalhamento is not None else ""
                    instituicao_finan = html.unescape(detalhamento.attrib.get("INSTITUICAO-FINANCIADORA", "")) if detalhamento is not None else ""

                    autores = trabalho_tecnico.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = trabalho_tecnico.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = trabalho_tecnico.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = trabalho_tecnico.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = trabalho_tecnico.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_trabalho.append((ano, titulo_trabalho_Tecnico, natureza, pais_Trabalho, meio_divulgacao,
                                                 Home_page, Finalidade, Disponibilidade, cidade, instituicao_finan,
                                                  nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores,  nome_completo ))

        ########################################## Processa demais_Apresentacao

        for demais_A in demais_Apresentacao:
            dados_basicos = demais_A.find("DADOS-BASICOS-DA-APRESENTACAO-DE-TRABALHO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_demais_A = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_A.find("DETALHAMENTO-DA-APRESENTACAO-DE-TRABALHO")
                    Nome_evento = html.unescape(detalhamento.attrib.get("NOME-DO-EVENTO", "")) if detalhamento is not None else ""
                    Insti_Promo = html.unescape(detalhamento.attrib.get("INSTITUICAO-PROMOTORA", "")) if detalhamento is not None else ""
                    local = html.unescape(detalhamento.attrib.get("LOCAL-DA-APRESENTACAO", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE-DA-APRESENTACAO", "")) if detalhamento is not None else ""

                    autores = demais_A.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_A.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_A.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_A.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_A.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_apresentacao.append((ano, titulo_demais_A, natureza, pais_Trabalho, meio_divulgacao, idioma, Nome_evento, Insti_Promo, local, cidade,nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores, nome_completo))

        ########################################## Processa demais_Materiais

        for demais_m in demais_Material:
            dados_basicos = demais_m.find("DADOS-BASICOS-DO-MATERIAL-DIDATICO-OU-INSTRUCIONAL")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Material = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    
                    detalhamento = demais_m.find("DETALHAMENTO-DO-MATERIAL-DIDATICO-OU-INSTRUCIONAL")
                    finalidade = html.unescape(detalhamento.attrib.get("FINALIDADE", "")) if detalhamento is not None else ""

                    autores = demais_m.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_m.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_m.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_m.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_m.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_material.append((ano, titulo, natureza, pais_Material, meio_divulgacao, idioma, finalidade,nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores, nome_completo))

########################################## Processa demais_editoracao

        for demais_E in demais_editoracao:
            dados_basicos = demais_E.find("DADOS-BASICOS-DE-EDITORACAO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_E.find("DETALHAMENTO-DE-EDITORACAO")
                    Numero_pag = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""
                    Insti_Promo = html.unescape(detalhamento.attrib.get("INSTITUICAO-PROMOTORA", "")) if detalhamento is not None else ""
                    Editora = html.unescape(detalhamento.attrib.get("EDITORA", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE", "")) if detalhamento is not None else ""

                    autores = demais_E.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_E.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_E.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_E.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_E.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_editoracao.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page, Numero_pag, Insti_Promo, Editora, cidade,nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas,descri_adc, setores,  nome_completo ))

########################################## Processa demais_evento

        for demais_ev in demais_evento:
            dados_basicos = demais_ev.find("DADOS-BASICOS-DA-ORGANIZACAO-DE-EVENTO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_ev.find("DETALHAMENTO-DA-ORGANIZACAO-DE-EVENTO")
                    Numero_pag = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""
                    Insti_Promo = html.unescape(detalhamento.attrib.get("INSTITUICAO-PROMOTORA", "")) if detalhamento is not None else ""
                    Editora = html.unescape(detalhamento.attrib.get("EDITORA", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE", "")) if detalhamento is not None else ""

                    autores = demais_ev.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_ev.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_ev.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_ev.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_ev.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_evento.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page, Numero_pag, Insti_Promo, Editora, cidade,nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas,descri_adc, setores,  nome_completo ))

########################################## Processa demais_programa

        for demais_P in demais_programa:
            dados_basicos = demais_P.find("DADOS-BASICOS-DO-PROGRAMA-DE-RADIO-OU-TV")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_P.find("DETALHAMENTO-DO-PROGRAMA-DE-RADIO-OU-TV")
                    Emissora = html.unescape(detalhamento.attrib.get("EMISSORA", "")) if detalhamento is not None else ""
                    Tema = html.unescape(detalhamento.attrib.get("TEMA", "")) if detalhamento is not None else ""
                    Data_apresentacao = html.unescape(detalhamento.attrib.get("DATA-DA-APRESENTACAO", "")) if detalhamento is not None else ""
                    Duracao_apresentacao_minutos = html.unescape(detalhamento.attrib.get("DURACAO-EM-MINUTOS", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE", "")) if detalhamento is not None else ""

                    autores = demais_P.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_P.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_P.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_P.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_P.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_programa.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page,
                                                        Emissora, Tema, Data_apresentacao,Duracao_apresentacao_minutos, cidade,nomes_autores, Nome_Citacao,  ordem_auditoria, 
                                                        chaves, areas,descri_adc, setores,  nome_completo ))

########################################## Processa demais_relatorio
        
        for demais_R in demais_relatorio:
            dados_basicos = demais_R.find("DADOS-BASICOS-DO-RELATORIO-DE-PESQUISA")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_R.find("DETALHAMENTO-DO-RELATORIO-DE-PESQUISA")
                    Nome_projeto = html.unescape(detalhamento.attrib.get("NOME-DO-PROJETO", "")) if detalhamento is not None else ""
                    numero_pags = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""
                    Disponibilidade = html.unescape(detalhamento.attrib.get("DISPONIBILIDADE", "")) if detalhamento is not None else ""
                    instituicao_finan = html.unescape(detalhamento.attrib.get("INSTITUICAO-FINANCIADORA", "")) if detalhamento is not None else ""

                    autores = demais_R.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_R.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_R.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_R.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_R.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_relatorio.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page,
                                                        Nome_projeto, numero_pags, Disponibilidade,instituicao_finan,nomes_autores, Nome_Citacao,  ordem_auditoria, 
                                                        chaves, areas,descri_adc, setores,  nome_completo ))
                    
########################################## Processa demais_midia

        for demais_Md in demais_midia:
            dados_basicos = demais_Md.find("DADOS-BASICOS-DA-MIDIA-SOCIAL-WEBSITE-BLOG")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_Md.find("DETALHAMENTO-DA-MIDIA-SOCIAL-WEBSITE-BLOG")
                    tema = html.unescape(detalhamento.attrib.get("TEMA", "")) if detalhamento is not None else ""

                    autores = demais_Md.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_Md.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_Md.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_Md.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_Md.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_midia.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page,
                                                        tema,nomes_autores, Nome_Citacao,  ordem_auditoria, 
                                                        chaves, areas,descri_adc, setores,  nome_completo ))
                    

########################################## Processa demais_outra
                    
        
        for demais_O in demais_outra:
            dados_basicos = demais_O.find("DADOS-BASICOS-DE-OUTRA-PRODUCAO-TECNICA")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_O.find("DETALHAMENTO-DE-OUTRA-PRODUCAO-TECNICA")
                    Finalidade = html.unescape(detalhamento.attrib.get("FINALIDADE", "")) if detalhamento is not None else ""
                    local = html.unescape(detalhamento.attrib.get("LOCAL", "")) if detalhamento is not None else ""
                    Cidade = html.unescape(detalhamento.attrib.get("CIDADE", "")) if detalhamento is not None else ""
                    instituicao_Promo = html.unescape(detalhamento.attrib.get("INSTITUICAO-PROMOTORA", "")) if detalhamento is not None else ""

                    autores = demais_O.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_O.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_O.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_O.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_O.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_outra.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page,
                                                        Finalidade, local, Cidade,instituicao_Promo,nomes_autores, Nome_Citacao,  ordem_auditoria, 
                                                        chaves, areas,descri_adc, setores,  nome_completo ))    
                    
########################################## Processa demais_carta

        for demais_C in demais_carta:
            dados_basicos = demais_C.find("DADOS-BASICOS-DE-CARTA-MAPA-OU-SIMILAR")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_C.find("DETALHAMENTO-DE-CARTA-MAPA-OU-SIMILAR")
                    Tema = html.unescape(detalhamento.attrib.get("TEMA-DA-CARTA-MAPA-OU-SIMILAR", "")) if detalhamento is not None else ""
                    tecnica = html.unescape(detalhamento.attrib.get("TECNICA-UTILIZADA", "")) if detalhamento is not None else ""
                    Finalidade = html.unescape(detalhamento.attrib.get("FINALIDADE", "")) if detalhamento is not None else ""
                    area_carta = html.unescape(detalhamento.attrib.get("AREA-REPRESENTADA", "")) if detalhamento is not None else ""
                    instituicao_finan = html.unescape(detalhamento.attrib.get("INSTITUICAO-FINANCIADORA", "")) if detalhamento is not None else ""

                    autores = demais_C.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_C.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_C.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_C.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_C.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_carta.append((ano, titulo, natureza, pais_Trabalho, meio_divulgacao, idioma, Home_page,
                                                        Tema,tecnica, Finalidade,area_carta, instituicao_finan,nomes_autores, Nome_Citacao,  ordem_auditoria, 
                                                        chaves, areas,descri_adc, setores,  nome_completo ))  
                    
########################################## Processa demais_curso

        for demais_Cur in demais_curso:
            dados_basicos = demais_Cur.find("DADOS-BASICOS-DE-CURSOS-CURTA-DURACAO-MINISTRADO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    nivel = html.unescape(dados_basicos.attrib.get("NIVEL-DO-CURSO", ""))
                    pais_Trabalho = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = demais_Cur.find("DETALHAMENTO-DE-CURSOS-CURTA-DURACAO-MINISTRADO")
                    Participacao_autores = html.unescape(detalhamento.attrib.get("PARTICIPACAO-DOS-AUTORES", "")) if detalhamento is not None else ""
                    Local = html.unescape(detalhamento.attrib.get("LOCAL-DO-CURSO", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE", "")) if detalhamento is not None else ""
                    duracao = html.unescape(detalhamento.attrib.get("DURACAO", "")) if detalhamento is not None else ""
                    unidade = html.unescape(detalhamento.attrib.get("UNIDADE", "")) if detalhamento is not None else ""
                    instituicao_Promo = html.unescape(detalhamento.attrib.get("INSTITUICAO-PROMOTORA-DO-CURSO", "")) if detalhamento is not None else ""

                    autores = demais_Cur.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = demais_Cur.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = demais_Cur.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = demais_Cur.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = demais_Cur.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_demais_curso.append((ano, titulo, nivel, pais_Trabalho, meio_divulgacao, idioma, Home_page,
                                                        Participacao_autores,Local, cidade,duracao,unidade, instituicao_Promo,nomes_autores, Nome_Citacao,  ordem_auditoria, 
                                                        chaves, areas,descri_adc, setores,  nome_completo )) 
                    

    except FileNotFoundError:
        print(f"O arquivo {arquivo_xml} não foi encontrado.")
    except ET.ParseError:
        print(f"Erro ao analisar o arquivo {arquivo_xml}.")
    except ValueError:
        print(f"Erro ao converter o ano para inteiro no arquivo {arquivo_xml}.")
    except Exception as e:
        print(f"Ocorreu um erro no arquivo {arquivo_xml}: {e}")
    return resultados_trabalho, resultados_demais_apresentacao, resultados_demais_material, resultados_demais_editoracao, resultados_demais_evento, resultados_demais_programa, resultados_demais_relatorio, resultados_demais_midia, resultados_demais_outra, resultados_demais_carta, resultados_demais_curso
```
