```python
import xml.etree.ElementTree as ET
import pandas as pd
import os
import html
```

```python
def processar_arquivo(arquivo_xml):
    try:
        # Tenta analisar o arquivo XML
        tree = ET.parse(arquivo_xml)
        root = tree.getroot()

        # Busca as informações gerais
        dados_gerais = root.find("DADOS-GERAIS")
        nome_completo = html.unescape(dados_gerais.attrib["NOME-COMPLETO"])

        print("Informações dos trabalhos encontrados (Ano >= 2021):")
        print(f"Nome: {nome_completo}")

        musicas = root.findall(".//MUSICA")
        orientacao_mestrado = root.findall(".//ORIENTACOES-CONCLUIDAS-PARA-MESTRADO")
        orientacao_doutorado = root.findall(".//ORIENTACOES-CONCLUIDAS-PARA-DOUTORADO")
        orientacao_posdoutorado = root.findall(".//ORIENTACOES-CONCLUIDAS-PARA-POS-DOUTORADO")
        orientacao_outras = root.findall(".//OUTRAS-ORIENTACOES-CONCLUIDAS")
        
        resultados_musica = []
        resultados_orientacao_mestrado = []
        resultados_orientacao_doutorado = []
        resultados_orientacao_posdoutorado = []
        resultados_orientacao_outras = []

###################### Processa musica

        for musica in musicas:
            dados_basicos = musica.find("DADOS-BASICOS-DA-MUSICA")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_musica = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    pais = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    meio_divulgacao = html.unescape(dados_basicos.attrib.get("MEIO-DE-DIVULGACAO", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE", ""))
                    
                    detalhamento = musica.find("DETALHAMENTO-DA-MUSICA")
                    tipo_de_evento = html.unescape(detalhamento.attrib.get("TIPO-DE-EVENTO", "")) if detalhamento is not None else ""
                    atividade_autores = html.unescape(detalhamento.attrib.get("ATIVIDADE-DOS-AUTORES", "")) if detalhamento is not None else ""
                    Formacao_instrumental = html.unescape(detalhamento.attrib.get("FORMACAO-INSTRUMENTAL", "")) if detalhamento is not None else ""
                    estreia = html.unescape(detalhamento.attrib.get("DATA-ESTREIA", "")) if detalhamento is not None else ""
                    local_estreia = html.unescape(detalhamento.attrib.get("LOCAL-DE-ESTREIA", "")) if detalhamento is not None else ""
                    duracao = html.unescape(detalhamento.attrib.get("DURACAO", "")) if detalhamento is not None else ""
                    inst_promo = html.unescape(detalhamento.attrib.get("INSTITUICAO-PROMOTORA-DO-EVENTO", "")) if detalhamento is not None else ""
                    cidade = html.unescape(detalhamento.attrib.get("CIDADE-DO-EVENTO", "")) if detalhamento is not None else ""

                    autores = musica.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = musica.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = musica.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = musica.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = musica.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_musica.append((ano, titulo_musica,pais, natureza, meio_divulgacao,idioma,
                                                 Home_page, tipo_de_evento,atividade_autores,Formacao_instrumental
                                                  ,estreia,local_estreia,duracao,inst_promo, cidade, 
                                                  nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc,
                                                    setores,  nome_completo ))

###################### Processa orientacao_mestrado

        for orientacao_m in orientacao_mestrado:
            dados_basicos = orientacao_m.find("DADOS-BASICOS-DE-ORIENTACOES-CONCLUIDAS-PARA-MESTRADO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_orientacao_m = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    pais_orientacao = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = orientacao_m.find("DETALHAMENTO-DE-ORIENTACOES-CONCLUIDAS-PARA-MESTRADO")
                    tipo = html.unescape(detalhamento.attrib.get("TIPO-DE-ORIENTACAO", "")) if detalhamento is not None else ""
                    nome_orientado = html.unescape(detalhamento.attrib.get("NOME-DO-ORIENTADO", "")) if detalhamento is not None else ""
                    codigo_instituicao = html.unescape(detalhamento.attrib.get("CODIGO-INSTITUICAO", "")) if detalhamento is not None else ""
                    nome_instituicao = html.unescape(detalhamento.attrib.get("NOME-DA-INSTITUICAO", "")) if detalhamento is not None else ""
                    codigo_curso = html.unescape(detalhamento.attrib.get("CODIGO-CURSO", "")) if detalhamento is not None else ""
                    nome_curso = html.unescape(detalhamento.attrib.get("NOME-DO-CURSO", "")) if detalhamento is not None else ""
                    bolsa = html.unescape(detalhamento.attrib.get("FLAG-BOLSA", "")) if detalhamento is not None else ""
                    nome_agencia = html.unescape(detalhamento.attrib.get("NOME-DA-AGENCIA", "")) if detalhamento is not None else ""
                    numero_id_orientado= html.unescape(detalhamento.attrib.get("NUMERO-ID-ORIENTADO", "")) if detalhamento is not None else ""
                    numero_paginas = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""

                    autores = orientacao_m.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = orientacao_m.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = orientacao_m.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = orientacao_m.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = orientacao_m.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_orientacao_mestrado.append((ano, titulo_orientacao_m, natureza, pais_orientacao, idioma,
                                                 Home_page, tipo, nome_orientado, codigo_instituicao, nome_instituicao,
                                                 codigo_curso,nome_curso,bolsa,nome_agencia,numero_id_orientado,numero_paginas,
                                                  nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores, 
                                                    nome_completo ))
                    
###################### Processa orientacao_doutorado

        for orientacao_d in orientacao_doutorado:
            dados_basicos = orientacao_d.find("DADOS-BASICOS-DE-ORIENTACOES-CONCLUIDAS-PARA-DOUTORADO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_orientacao_d = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    pais_orientacao = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = orientacao_d.find("DETALHAMENTO-DE-ORIENTACOES-CONCLUIDAS-PARA-DOUTORADO")
                    tipo = html.unescape(detalhamento.attrib.get("TIPO-DE-ORIENTACAO", "")) if detalhamento is not None else ""
                    nome_orientado = html.unescape(detalhamento.attrib.get("NOME-DO-ORIENTADO", "")) if detalhamento is not None else ""
                    codigo_instituicao = html.unescape(detalhamento.attrib.get("CODIGO-INSTITUICAO", "")) if detalhamento is not None else ""
                    nome_instituicao = html.unescape(detalhamento.attrib.get("NOME-DA-INSTITUICAO", "")) if detalhamento is not None else ""
                    codigo_curso = html.unescape(detalhamento.attrib.get("CODIGO-CURSO", "")) if detalhamento is not None else ""
                    nome_curso = html.unescape(detalhamento.attrib.get("NOME-DO-CURSO", "")) if detalhamento is not None else ""
                    bolsa = html.unescape(detalhamento.attrib.get("FLAG-BOLSA", "")) if detalhamento is not None else ""
                    nome_agencia = html.unescape(detalhamento.attrib.get("NOME-DA-AGENCIA", "")) if detalhamento is not None else ""
                    numero_id_orientado= html.unescape(detalhamento.attrib.get("NUMERO-ID-ORIENTADO", "")) if detalhamento is not None else ""
                    numero_paginas = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""

                    autores = orientacao_d.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = orientacao_d.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = orientacao_d.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = orientacao_d.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = orientacao_d.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_orientacao_doutorado.append((ano, titulo_orientacao_d, natureza, pais_orientacao, idioma,
                                                 Home_page, tipo, nome_orientado, codigo_instituicao, nome_instituicao,
                                                 codigo_curso,nome_curso,bolsa,nome_agencia,numero_id_orientado,numero_paginas,
                                                  nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores, 
                                                    nome_completo ))
                    
###################### Processa orientacao_posdoutorado

        for orientacao_pd in orientacao_posdoutorado:
            dados_basicos = orientacao_pd.find("DADOS-BASICOS-DE-ORIENTACOES-CONCLUIDAS-PARA-POS-DOUTORADO")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_orientacao_pd = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    pais_orientacao = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = orientacao_pd.find("DETALHAMENTO-DE-ORIENTACOES-CONCLUIDAS-PARA-POS-DOUTORADO")
                    tipo = html.unescape(detalhamento.attrib.get("TIPO-DE-ORIENTACAO", "")) if detalhamento is not None else ""
                    nome_orientado = html.unescape(detalhamento.attrib.get("NOME-DO-ORIENTADO", "")) if detalhamento is not None else ""
                    codigo_instituicao = html.unescape(detalhamento.attrib.get("CODIGO-INSTITUICAO", "")) if detalhamento is not None else ""
                    nome_instituicao = html.unescape(detalhamento.attrib.get("NOME-DA-INSTITUICAO", "")) if detalhamento is not None else ""
                    codigo_curso = html.unescape(detalhamento.attrib.get("CODIGO-CURSO", "")) if detalhamento is not None else ""
                    nome_curso = html.unescape(detalhamento.attrib.get("NOME-DO-CURSO", "")) if detalhamento is not None else ""
                    bolsa = html.unescape(detalhamento.attrib.get("FLAG-BOLSA", "")) if detalhamento is not None else ""
                    nome_agencia = html.unescape(detalhamento.attrib.get("NOME-DA-AGENCIA", "")) if detalhamento is not None else ""
                    numero_id_orientado= html.unescape(detalhamento.attrib.get("NUMERO-ID-ORIENTADO", "")) if detalhamento is not None else ""
                    numero_paginas = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""


                    autores = orientacao_pd.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = orientacao_pd.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = orientacao_pd.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = orientacao_pd.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = orientacao_pd.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_orientacao_posdoutorado.append((ano, titulo_orientacao_pd, natureza, pais_orientacao, idioma,
                                                 Home_page, tipo, nome_orientado, codigo_instituicao, nome_instituicao,
                                                 codigo_curso,nome_curso,bolsa,nome_agencia,numero_id_orientado,numero_paginas,
                                                  nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores, 
                                                    nome_completo ))

###################### Processa orientacao_outras

        for orientacao_o in orientacao_outras:
            dados_basicos = orientacao_o.find("DADOS-BASICOS-DE-OUTRAS-ORIENTACOES-CONCLUIDAS")     
            if dados_basicos is not None:
                ano = dados_basicos.attrib.get("ANO", None)
                if ano is not None and int(ano) >= 2021:
                    titulo_orientacao_o = html.unescape(dados_basicos.attrib.get("TITULO", ""))
                    pais_orientacao = html.unescape(dados_basicos.attrib.get("PAIS", ""))
                    idioma = html.unescape(dados_basicos.attrib.get("IDIOMA", ""))
                    natureza = html.unescape(dados_basicos.attrib.get("NATUREZA", ""))
                    Home_page = html.unescape(dados_basicos.attrib.get("HOME-PAGE-DO-TRABALHO", ""))
                    
                    detalhamento = orientacao_o.find("DETALHAMENTO-DE-OUTRAS-ORIENTACOES-CONCLUIDAS")
                    tipo = html.unescape(detalhamento.attrib.get("TIPO-DE-ORIENTACAO-CONCLUIDA", "")) if detalhamento is not None else ""
                    nome_orientado = html.unescape(detalhamento.attrib.get("NOME-DO-ORIENTADO", "")) if detalhamento is not None else ""
                    codigo_instituicao = html.unescape(detalhamento.attrib.get("CODIGO-INSTITUICAO", "")) if detalhamento is not None else ""
                    nome_instituicao = html.unescape(detalhamento.attrib.get("NOME-DA-INSTITUICAO", "")) if detalhamento is not None else ""
                    codigo_curso = html.unescape(detalhamento.attrib.get("CODIGO-CURSO", "")) if detalhamento is not None else ""
                    nome_curso = html.unescape(detalhamento.attrib.get("NOME-DO-CURSO", "")) if detalhamento is not None else ""
                    bolsa = html.unescape(detalhamento.attrib.get("FLAG-BOLSA", "")) if detalhamento is not None else ""
                    nome_agencia = html.unescape(detalhamento.attrib.get("NOME-DA-AGENCIA", "")) if detalhamento is not None else ""
                    numero_id_orientado= html.unescape(detalhamento.attrib.get("NUMERO-ID-ORIENTADO", "")) if detalhamento is not None else ""
                    numero_paginas = html.unescape(detalhamento.attrib.get("NUMERO-DE-PAGINAS", "")) if detalhamento is not None else ""

                    autores = orientacao_o.findall("AUTORES")
                    nomes_autores = [html.unescape(autor.attrib.get("NOME-COMPLETO-DO-AUTOR", "")) for autor in autores]
                    Nome_Citacao = [html.unescape(autor.attrib.get("NOME-PARA-CITACAO", "")) for autor in autores]
                    ordem_auditoria = [html.unescape(autor.attrib.get("ORDEM-DE-AUTORIA", "")) for autor in autores]
                    
                    Palavra_chaves = orientacao_o.find("PALAVRAS-CHAVE")
                    chaves = []
                    if Palavra_chaves is not None:
                        for chave in Palavra_chaves.attrib:
                            if chave.startswith("PALAVRA-CHAVE-"):
                                valor_chave = html.unescape(Palavra_chaves.attrib[chave])
                                if valor_chave:
                                    chaves.append(valor_chave)

                    # Extração das áreas do conhecimento
                    areas_do_conhecimento = orientacao_o.find("AREAS-DO-CONHECIMENTO")
                    areas = []
                    if areas_do_conhecimento is not None:
                        for area in areas_do_conhecimento:
                            nome_grande_area = html.unescape(area.attrib.get("NOME-GRANDE-AREA-DO-CONHECIMENTO", ""))
                            nome_area = html.unescape(area.attrib.get("NOME-DA-AREA-DO-CONHECIMENTO", ""))
                            nome_sub_area = html.unescape(area.attrib.get("NOME-DA-SUB-AREA-DO-CONHECIMENTO", ""))
                            nome_especialidade = html.unescape(area.attrib.get("NOME-DA-ESPECIALIDADE", ""))
                            areas.append((nome_grande_area, nome_area, nome_sub_area, nome_especialidade))

                    infoadc = orientacao_o.findall("INFORMACOES-ADICIONAIS")
                    descri_adc = [html.unescape(info.attrib.get("DESCRICAO-INFORMACOES-ADICIONAIS", "")) for info in infoadc]

                    Setores_atividade = orientacao_o.find("SETORES-DE-ATIVIDADE")
                    setores = []
                    if Setores_atividade is not None:
                        for setor in Setores_atividade.attrib:
                            if setor.startswith("SETOR-DE-ATIVIDADE-"):
                                valor_Setor = html.unescape(Setores_atividade.attrib[setor])
                                if valor_Setor:
                                    setores.append(valor_Setor)
                    
                    resultados_orientacao_outras.append((ano, titulo_orientacao_o, natureza, pais_orientacao, idioma,
                                                 Home_page, tipo, nome_orientado, codigo_instituicao, nome_instituicao,
                                                 codigo_curso,nome_curso,bolsa,nome_agencia,numero_id_orientado,numero_paginas,
                                                  nomes_autores, Nome_Citacao,  ordem_auditoria, chaves, areas, descri_adc, setores, 
                                                    nome_completo ))
                    

    except FileNotFoundError:
        print(f"O arquivo {arquivo_xml} não foi encontrado.")
    except ET.ParseError:
        print(f"Erro ao analisar o arquivo {arquivo_xml}.")
    except ValueError:
        print(f"Erro ao converter o ano para inteiro no arquivo {arquivo_xml}.")
    except Exception as e:
        print(f"Ocorreu um erro no arquivo {arquivo_xml}: {e}")
        return [], [], [], [], [], [], [], []
    return resultados_musica, resultados_orientacao_mestrado , resultados_orientacao_doutorado ,resultados_orientacao_posdoutorado ,resultados_orientacao_outras 
    
```
